```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
# convert to datetime
df['hour'] = pd.to_datetime(df['hour'])
df['hour_of_day'] = df['hour'].dt.hour
df['day_of_week'] = df['hour'].dt.dayofweek
```

```python
device_id_counts = df['device_id'].value_counts()
df['device_id_counts'] = df['device_id'].map(device_id_counts)
```

```python
device_ip_counts = df['device_id'].value_counts()
df['device_id_counts'] = df['device_id'].map(device_ip_counts)
```

```python
hourly_counts = df.groupby('hour_of_day')['device_ip'].count()
df['hourly_user_ip_count'] = df['hour_of_day'].map(hourly_counts)
```

```python
weekly_counts = df.groupby('day_of_week')['device_ip'].count()
df['weekly_user_id_counts'] = df['day_of_week'].map(weekly_counts)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
# click distribution
df['click'].value_counts().to_dict()
```

```python
# count unique per column
df.nunique().to_dict()
```

```python
# value count of specific column
df['device_conn_type'].value_counts().to_dict()
```

```python
# id vs device_type
df['device_conn_type'].value_counts().to_dict()
```

```python
# heatmap
df.select_dtypes('number').corr().to_dict()
```

```python
group_data = df.groupby(['banner_pos', 'click']).size().unstack()
group_data['Frequency'] = df['banner_pos'].value_counts(normalize=True)
ctr = group_data[1] / (group_data[0] + group_data[1])
group_data['CTR'] = ctr
group_data = group_data.fillna(0)
```

```python
group_data.to_dict()
```

```python
device_id_counts = df['device_id'].value_counts()
df['device_id_counts'] = df['device_id'].map(device_id_counts)
# device_IP_counts
```

```python
# convert to datetime
df['hour'] = pd.to_datetime(df['hour'])
```

```python
df['hour_of_day'] = df['hour'].dt.hour
df['day_of_week'] = df['hour'].dt.dayofweek
```

```python
df[['hour', 'click']].groupby('hour').agg({'click': 'sum'}).to_dict()
```
